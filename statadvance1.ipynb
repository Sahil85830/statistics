```python
#1. What is a random variable in probability theory?
'''A **random variable** in probability theory is a function that assigns numerical values to outcomes of a random experiment. It can be:  

- **Discrete**: Takes a finite or countable set of values (e.g., number of heads in coin flips).  
- **Continuous**: Takes values from an interval or range (e.g., height of people in a population).  

Random variables help quantify uncertainty and are characterized by probability distributions.'''
```

```python
#2. What are the types of random variables?
'''Random variables are classified into two main types:  

1. **Discrete Random Variable**  
   - Takes a finite or countable number of values.  
   - Example: Number of heads in 5 coin flips.  
   - Described by a **probability mass function (PMF)**.  

2. **Continuous Random Variable**  
   - Takes an uncountable number of values within a range.  
   - Example: Height of students in a class.  
   - Described by a **probability density function (PDF)**.  '''
```

```python
#3. What is the difference between discrete and continuous distributions?
'''The key differences between **discrete** and **continuous** distributions are:  

| Feature                           | Discrete Distribution                                       | Continuous Distribution                        
|----------------------             |--------------------------------                             |--------------------------------
| **Definition**                    | Probability distribution of a **discrete** random variable. | Probability distribution of a **continuous** random variable. 
| **Values Taken**                  | Finite or countable values.                                 | Infinite or uncountable values in an interval. 
| **Probability Representation**    | Probability Mass Function (PMF).                            | Probability Density Function (PDF). 
| **Probability of a Single Value** | \( P(X = x) \) has a positive probability.                  | \( P(X = x) = 0 \); probability is given over an interval. 
| **Example**                       | Number of heads in 5 coin flips.                            | Heights of students in a school. '''
```

```python
#4. What are probability distribution functions (PDF)?
'''A **Probability Distribution Function (PDF)** describes the likelihood of a random variable taking specific values. It differs based on the type of random variable:  

1. **For Discrete Random Variables**  
   - Uses a **Probability Mass Function (PMF)**, which gives the probability that the variable takes a specific value.  
   - Example: Probability of rolling a three on a fair die.  

2. **For Continuous Random Variables**  
   - Uses a **Probability Density Function (PDF)**, where probabilities are found over intervals rather than specific values.  
   - The probability of a single value is zero, so we calculate the probability over a range using integration.  
   - Example: Distribution of people's heights in a population.'''
```

```python
#5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
'''The key differences between **Cumulative Distribution Functions (CDF)** and **Probability Distribution Functions (PDF)** are:  

1. **Definition**  
   - PDF (or PMF for discrete cases) gives the probability of a random variable taking a specific value or range.  
   - CDF gives the probability that a random variable is less than or equal to a given value.  

2. **Formula**  
   - For a discrete variable, the CDF is the sum of probabilities from the PMF.  
   - For a continuous variable, the CDF is the integral of the PDF.  

3. **Probability Representation**  
   - PDF/PMF provides probabilities of individual values or small intervals.  
   - CDF provides cumulative probabilities up to a given point.  

4. **Example**  
   - PDF: The probability of rolling a three on a fair die is one by six.  
   - CDF: The probability of rolling a number less than or equal to three is three by six.'''
```

```python
#6. What is a discrete uniform distribution?
'''A **Discrete Uniform Distribution** is a probability distribution where a finite number of outcomes are equally likely.  

### **Key Features:**  
1. **Equal Probability**: Each outcome has the same probability.  
2. **Probability Mass Function (PMF)**: Given by  
   - P(X = x) = 1 / n, where **n** is the total number of possible outcomes.  
3. **Mean (Expected Value)**: (n + 1) / 2  
4. **Variance**: (nÂ² - 1) / 12  

### **Example:**  
Rolling a **fair six-sided die** follows a discrete uniform distribution since each face (1 to 6) has an equal probability of one by six.'''
```

```python
#7. What are the key properties of a Bernoulli distribution?
'''The **Bernoulli distribution** is a discrete probability distribution for a random experiment with only two possible outcomes: **success (1)** or **failure (0)**.  

### **Key Properties:**  
1. **Single Trial**: Represents a single experiment with two outcomes.  
2. **Probability Mass Function (PMF)**:  
   - P(X = 1) = p (probability of success)  
   - P(X = 0) = 1 - p (probability of failure)  
3. **Mean (Expected Value)**: E(X) = p  
4. **Variance**: Var(X) = p(1 - p)  
5. **Moment-Generating Function (MGF)**: M(t) = 1 - p + p * e^t  
6. **Special Case**: A Binomial distribution with **n = 1** is a Bernoulli distribution.  

### **Example:**  
Flipping a fair coin, where heads (success) occurs with probability 0.5 and tails (failure) occurs with probability 0.5. '''
```

```python
#8. What is the binomial distribution, and how is it used in probability?
'''The **Binomial Distribution** is a discrete probability distribution that models the number of successes in a fixed number of independent **Bernoulli trials**, each with the same probability of success.  

### **Key Features:**  
1. **Fixed Number of Trials (n)**: The experiment is repeated **n** times.  
2. **Two Possible Outcomes**: Success (1) or Failure (0) in each trial.  
3. **Constant Probability (p)**: The probability of success remains the same in each trial.  
4. **Independent Trials**: Each trial does not affect the others.  

### **Probability Mass Function (PMF):**  
P(X = k) = (n choose k) * p^k * (1 - p)^(n - k),  
where **k** is the number of successes.  

### **Mean and Variance:**  
- Mean (Expected Value): E(X) = n * p  
- Variance: Var(X) = n * p * (1 - p)  

### **Usage in Probability:**  
- **Coin Tosses**: Probability of getting exactly 3 heads in 5 flips.  
- **Quality Control**: Probability of finding defective items in a batch.  
- **Medical Studies**: Probability of a certain number of patients recovering after treatment.'''
```

```python
#9. What is the Poisson distribution and where is it applied?
'''The **Poisson distribution** is a discrete probability distribution that models the number of events occurring in a fixed interval of time or space, given that the events occur independently and at a constant average rate.  

### **Key Features:**  
1. **Describes Rare Events**: Used for counting occurrences of rare events in a large sample.  
2. **Probability Mass Function (PMF):**  
   P(X = k) = (lambda^k * e^(-lambda)) / k!  
   where **lambda** is the average rate of occurrences, and **k** is the number of events.  
3. **Mean and Variance:**  
   - Mean (Expected Value): E(X) = lambda  
   - Variance: Var(X) = lambda  

### **Applications:**  
- **Queueing Systems**: Number of customers arriving at a service center per hour.  
- **Network Traffic**: Number of data packets arriving at a server in a second.  
- **Biology**: Number of mutations in a DNA strand over time.  
- **Call Centers**: Number of calls received in a given time period.'''
```

```python
#10.  What is a continuous uniform distribution?
'''A **Continuous Uniform Distribution** is a probability distribution in which all values within a given interval are equally likely. It is defined by two parameters: **a** (minimum value) and **b** (maximum value).  

### **Key Features:**  
1. **Equal Probability**: Every value in the range [a, b] has the same probability density.  
2. **Probability Density Function (PDF):**  
   - f(x) = 1 / (b - a), for a â¤ x â¤ b  
   - f(x) = 0, otherwise  
3. **Cumulative Distribution Function (CDF):**  
   - F(x) = (x - a) / (b - a), for a â¤ x â¤ b  
4. **Mean (Expected Value):** E(X) = (a + b) / 2  
5. **Variance:** Var(X) = (b - a)Â² / 12  

### **Applications:**  
- **Random Number Generation**: Simulating fair random values between two limits.  
- **Manufacturing**: Modeling uniform thickness of materials.  
- **Physics**: Estimating arrival times when all times in an interval are equally likely.'''
```

```python
#11. What are the characteristics of a normal distribution?
'''The **Normal Distribution**, also known as the **Gaussian distribution**, is a continuous probability distribution that is symmetric and bell-shaped. It is widely used in statistics and natural phenomena.  

### **Key Characteristics:**  
1. **Bell-Shaped Curve**: Symmetric around the mean.  
2. **Defined by Two Parameters**:  
   - **Mean (Î¼):** The central location of the distribution.  
   - **Standard Deviation (Ï):** Measures the spread of data.  
3. **Probability Density Function (PDF):**  
   - f(x) = (1 / (Ïâ2Ï)) * exp(- (x - Î¼)Â² / (2ÏÂ²))  
4. **68-95-99.7 Rule (Empirical Rule):**  
   - 68% of values lie within **1Ï** of the mean.  
   - 95% within **2Ï**.  
   - 99.7% within **3Ï**.  
5. **Mean = Median = Mode**: The peak of the curve is at **Î¼**.  
6. **Asymptotic to the X-axis**: The tails approach but never touch the axis.  
7. **Used in Central Limit Theorem (CLT)**: Sum of many independent random variables tends to be normally distributed.  

### **Applications:**  
- **Natural Phenomena**: Heights, weights, IQ scores.  
- **Finance**: Stock market returns.  
- **Quality Control**: Measuring variations in manufacturing processes.'''
```

```python
#12. What is the standard normal distribution, and why is it important?
'''The **Standard Normal Distribution** is a special case of the normal distribution where the **mean is zero** and the **standard deviation is one**. It is represented by the random variable **Z** and follows a bell-shaped curve.  

### **Importance of the Standard Normal Distribution:**  
1. **Simplifies Calculations**: Converts any normal distribution to a common scale using the **Z-score** formula, which is calculated as the value minus the mean divided by the standard deviation.  
2. **Universal Reference**: Probability values are precomputed in **Z-tables**, making it easy to find probabilities.  
3. **Used in Hypothesis Testing**: Helps in **Z-tests** for comparing sample means.  
4. **Integral to Confidence Intervals**: Used in determining confidence levels for population estimates.  
5. **Foundation for Central Limit Theorem**: Many real-world distributions approximate a normal distribution when sample sizes are large.  

### **Example Application:**  
Finding the probability that a test score falls below a certain value in a normally distributed dataset by converting it to a Z-score and using standard normal tables.'''
```

```python
#13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
'''The **Central Limit Theorem (CLT)** states that the sampling distribution of the sample mean approaches a **normal distribution**, regardless of the population's original distribution, as the **sample size increases**.  

### **Key Aspects of CLT:**  
1. **Normal Approximation**: For large samples, the distribution of the sample mean becomes approximately normal.  
2. **Mean and Standard Deviation**:  
   - The mean of the sample means is equal to the population mean.  
   - The standard deviation of the sample means, called the **standard error**, is the population standard deviation divided by the square root of the sample size.  
3. **Sample Size Rule**: A sample size of **30 or more** is generally considered sufficient for CLT to hold.  

### **Importance of CLT in Statistics:**  
1. **Justifies Use of Normal Distribution**: Allows statistical tests like **Z-tests and t-tests** to be applied even if the original data is not normal.  
2. **Supports Inferential Statistics**: Helps in estimating population parameters from sample data.  
3. **Used in Confidence Intervals and Hypothesis Testing**: Ensures valid conclusions from sample data.  
4. **Applies to Many Real-World Situations**: Used in economics, quality control, polling, and machine learning.  

### **Example Application:**  
In election polling, the average response from a **random sample** of voters follows a normal distribution, enabling accurate predictions about the entire population.'''
```

```python
#14. How does the Central Limit Theorem relate to the normal distribution?
'''The **Central Limit Theorem (CLT)** relates to the **normal distribution** by stating that, as the sample size increases, the distribution of the **sample mean** approaches a normal distribution, regardless of the shape of the original population distribution.  

### **Key Connections Between CLT and Normal Distribution:**  
1. **Normal Approximation of Sample Means**:  
   - If we take repeated random samples from any population, the distribution of their means will be approximately **normal** for large sample sizes.  
2. **Standard Error and Spread Reduction**:  
   - The standard deviation of the sample means, called the **standard error**, decreases as the sample size increases, making the distribution narrower.  
3. **Foundation for Statistical Inference**:  
   - Since the sample mean follows a normal distribution, we can use **Z-tests, t-tests, and confidence intervals** to make conclusions about the population.  

### **Practical Example:**  
- Even if individual IQ scores in a population are **not normally distributed**, the **average IQ** of randomly selected groups of people will follow a **normal distribution** when the sample size is large.'''
```

```python
#15. What is the application of Z statistics in hypothesis testing?
'''### **Application of Z-Statistics in Hypothesis Testing**  

**Z-statistics** are used in hypothesis testing when the population standard deviation is known and the sample size is large (**n â¥ 30**). It helps determine whether a sample mean significantly differs from the population mean.  

### **Key Applications:**  
1. **One-Sample Z-Test:**  
   - Tests whether the sample mean differs from the population mean.  
   - Example: Checking if the average height of students in a school differs from the national average.  

2. **Two-Sample Z-Test:**  
   - Compares the means of two independent samples.  
   - Example: Comparing the average test scores of students from two different schools.  

3. **Proportion Z-Test:**  
   - Tests hypotheses about population proportions.  
   - Example: Determining if the proportion of voters supporting a candidate differs from 50 percent.  

### **Decision Rule Using Z-Score:**  
- Calculate the **Z-score** using the formula:  
  **Z = (Sample Mean - Population Mean) / (Standard Deviation / sqrt(n))**  
- Compare it with the critical value from the **Z-table** based on the significance level (**alpha**).  
- If **Z falls in the rejection region**, reject the null hypothesis.  

### **Practical Use:**  
- Quality control in manufacturing.  
- Medical research to test drug effectiveness.  
- Market research for consumer behavior analysis.'''
```

```python
#16. How do you calculate a Z-score, and what does it represent?
'''### **Calculation of a Z-Score**  
A **Z-score** is calculated using the formula:  

**Z = (X - Î¼) / Ï**  

where:  
- **X** = observed value  
- **Î¼** = population mean  
- **Ï** = population standard deviation  

For a sample, if the population standard deviation is unknown, the formula is:  

**Z = (X - Î¼) / (Ï / sqrt(n))**  

where **n** is the sample size.  

### **What a Z-Score Represents**  
1. **Standardized Value**: A Z-score indicates how many **standard deviations** an observation is from the **mean**.  
2. **Comparison Across Distributions**: Converts values from different distributions to a common standard.  
3. **Probability Interpretation**:  
   - A **Z-score of 0** means the value is equal to the mean.  
   - A **positive Z-score** means the value is above the mean.  
   - A **negative Z-score** means the value is below the mean.  

### **Example Calculation**  
If the average test score is **70** with a standard deviation of **10**, and a student scores **85**:  
**Z = (85 - 70) / 10 = 1.5**  
This means the student scored **1.5 standard deviations above the mean**.'''
```

```python
#17. What are point estimates and interval estimates in statistics?
'''### **Point Estimates and Interval Estimates in Statistics**  

#### **Point Estimates**  
A **point estimate** is a single value used to approximate an unknown population parameter. It provides a best guess but does not indicate uncertainty.  

For example, the sample mean is a point estimate of the population mean. Other common point estimates include the sample proportion for the population proportion and the sample standard deviation for the population standard deviation.  

#### **Interval Estimates**  
An **interval estimate** provides a range of values within which the population parameter is likely to fall, with a certain level of confidence.  

For example, a ninety five percent confidence interval for a population mean might be fifty to sixty, meaning the true mean is likely within this range.  

The formula for a confidence interval depends on whether the population standard deviation is known or unknown.  

When the population standard deviation is known, the confidence interval is calculated as the sample mean plus or minus the Z-score multiplied by the population standard deviation divided by the square root of the sample size.  

When the population standard deviation is unknown, the confidence interval is calculated as the sample mean plus or minus the t-score multiplied by the sample standard deviation divided by the square root of the sample size.  

#### **Key Differences**  
A point estimate provides a single value, while an interval estimate gives a range of values. A point estimate is less accurate because it does not account for variability, whereas an interval estimate is more reliable since it considers uncertainty.  

For example, a point estimate might state that the sample mean is fifty, whereas an interval estimate might state that the confidence interval is between forty five and fifty five.  

#### **Use in Statistics**  
Point estimates provide a quick, single-value approximation, while interval estimates give a more reliable range that incorporates uncertainty.'''
```

```python
#18. What is the significance of confidence intervals in statistical analysis?
'''### **Significance of Confidence Intervals in Statistical Analysis**  

A **confidence interval** is a range of values that estimates an unknown population parameter with a certain level of confidence. It is widely used in statistics to measure the reliability of an estimate.  

#### **Key Significance**  

1. **Measures Uncertainty**  
   Confidence intervals account for variability in sample data, providing a more accurate estimate than a single point estimate.  

2. **Provides a Range Instead of a Single Value**  
   Unlike point estimates, confidence intervals show the possible range within which the true population parameter is likely to fall.  

3. **Helps in Hypothesis Testing**  
   If a confidence interval does not contain a hypothesized value, the null hypothesis may be rejected.  

4. **Used for Decision Making**  
   In business, medicine, and social sciences, confidence intervals help make informed decisions based on sample data.  

5. **Determines Precision of Estimates**  
   A narrower confidence interval indicates a more precise estimate, while a wider interval suggests greater uncertainty.  

#### **Example**  
If a study finds that the average test score of students is seventy with a ninety five percent confidence interval of sixty six to seventy four, it means there is a ninety five percent chance that the true average score lies within this range.'''
```

```python
#19. What is the relationship between a Z-score and a confidence interval?
'''### **Relationship Between a Z-Score and a Confidence Interval**  

A **Z-score** and a **confidence interval** are closely related in statistical analysis. The Z-score is used to determine the margin of error in confidence interval calculations when the population standard deviation is known.  

#### **Key Relationships**  

1. **Z-Score Determines Confidence Level**  
   The Z-score corresponds to the probability that a value falls within a certain range in a standard normal distribution. Common Z-scores for confidence levels are:  
   - **90 percent confidence level** â Z = 1.645  
   - **95 percent confidence level** â Z = 1.96  
   - **99 percent confidence level** â Z = 2.576  

2. **Confidence Interval Formula Using Z-Score**  
   When the population standard deviation is known, the confidence interval for the population mean is calculated as:  
   **Confidence Interval = Sample Mean Â± Z * (Population Standard Deviation / sqrt(Sample Size))**  

3. **Wider Confidence Intervals for Higher Confidence Levels**  
   A larger Z-score results in a wider confidence interval, reflecting greater certainty that the true population parameter falls within the range.  

#### **Example**  
If a sample mean is fifty with a standard deviation of five and a sample size of thirty, the ninety five percent confidence interval is calculated as:  
Fifty plus or minus one point nine six multiplied by five divided by the square root of thirty.  

This interval provides a range in which the true population mean is likely to fall with ninety five percent confidence.'''
```

```python
#20. How are Z-scores used to compare different distributions?
'''### **How Z-Scores Are Used to Compare Different Distributions**  

A **Z-score** standardizes data from different distributions, allowing for direct comparison even if the original scales and units are different.  

#### **Key Uses of Z-Scores for Comparison**  

1. **Standardization of Different Distributions**  
   - Converts values from any normal distribution to a standard normal distribution with a mean of zero and a standard deviation of one.  
   - The formula **Z = (X - Î¼) / Ï** makes different datasets comparable.  

2. **Identifies Relative Position**  
   - A Z-score indicates how far a data point is from the mean in terms of standard deviations.  
   - A higher Z-score means the value is further above the mean, while a lower Z-score means it is below the mean.  

3. **Comparison Across Different Units and Scales**  
   - Allows comparison of test scores from different exams, heights from different populations, or performance in different fields.  
   - Example: Comparing SAT and ACT scores by converting them into Z-scores.  

4. **Determining Outliers and Extremes**  
   - Data points with very high or very low Z-scores (e.g., beyond Â±3) are considered outliers.  

#### **Example**  
If Student A scores 85 on a test with a mean of 75 and a standard deviation of 5, and Student B scores 90 on another test with a mean of 80 and a standard deviation of 6:  
- Z-score for Student A = (85 - 75) / 5 = 2  
- Z-score for Student B = (90 - 80) / 6 = 1.67  

Even though Student B has a higher raw score, Student A performed better relative to their group.'''
```

```python
#21. What are the assumptions for applying the Central Limit Theorem?
'''### **Assumptions for Applying the Central Limit Theorem (CLT)**  

The **Central Limit Theorem (CLT)** states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's original distribution. However, certain assumptions must be met for CLT to hold effectively.  

#### **Key Assumptions:**  

1. **Random Sampling**  
   - The samples must be randomly selected from the population to ensure an unbiased representation.  

2. **Independence of Observations**  
   - Each observation must be independent of the others.  
   - This is satisfied when sampling with replacement or when the sample size is less than 10 percent of the population in cases of sampling without replacement.  

3. **Sample Size is Sufficiently Large**  
   - For most distributions, a sample size of **30 or more** is considered sufficient for CLT to apply.  
   - If the population distribution is highly skewed, a larger sample size may be needed.  

4. **Finite Variance**  
   - The population from which samples are drawn should have a finite standard deviation. Extremely heavy-tailed distributions may violate this assumption.  

#### **Practical Importance:**  
- Ensures that statistical inference methods like confidence intervals and hypothesis tests can be applied using normal approximations.  
- Helps in analyzing real-world data, even when the original population is not normally distributed.'''
```

```python
#22. What is the concept of expected value in a probability distribution?
'''### **Concept of Expected Value in a Probability Distribution**  

The **expected value** of a probability distribution represents the long-term average or mean outcome of a random variable when an experiment is repeated many times. It is a measure of the central tendency of the distribution.  

#### **Formula for Expected Value (E[X]):**  

1. **For a Discrete Random Variable:**  
   **E[X] = Sum of (X_i * P(X_i))**  
   - Here, **X_i** represents possible values of the random variable, and **P(X_i)** is their corresponding probabilities.  

2. **For a Continuous Random Variable:**  
   **E[X] = Integral of (X * f(X) dX)**  
   - Here, **f(X)** is the probability density function (PDF) of the continuous variable.  

#### **Interpretation of Expected Value:**  
- It does not predict a single outcome but provides the average value over many trials.  
- It is used in decision-making, gambling, finance, and risk analysis.  

#### **Example:**  
In a fair six-sided die roll, the expected value is:  
**E[X] = (1 * 1/6) + (2 * 1/6) + (3 * 1/6) + (4 * 1/6) + (5 * 1/6) + (6 * 1/6) = 3.5**  
This means that, on average, rolling the die many times will result in an outcome close to 3.5.'''
```

```python
#23. How does a probability distribution relate to the expected outcome of a random variable?
'''A **probability distribution** describes all possible outcomes of a random variable and their corresponding probabilities. It gives the likelihood of each possible value the random variable can take. The **expected value** of a random variable is a specific measure derived from the probability distribution.  

### **Relationship Between Probability Distribution and Expected Value:**  

1. **Expected Value as the Weighted Average**  
   The expected value is the **weighted average** of all possible outcomes, with each outcome weighted by its probability. In other words, it is calculated by multiplying each possible outcome by its probability and summing the results.  

2. **Central Tendency of the Distribution**  
   The expected value represents the **central point** of the probability distribution. It is the value around which the random variable is likely to cluster in the long run.  

3. **Use of Probability Distribution to Calculate Expected Value**  
   - For a **discrete random variable**, the expected value is calculated by multiplying each outcome by its probability and summing the products.  
   - For a **continuous random variable**, the expected value is calculated using the integral of the product of the variableâs value and its probability density function.  

### **Example:**  
If you have a discrete random variable **X** representing the outcomes of a fair die roll with outcomes 1 through 6, the probability distribution assigns each outcome a probability of 1/6. The expected value of **X** is calculated as:  

**E[X] = (1 * 1/6) + (2 * 1/6) + (3 * 1/6) + (4 * 1/6) + (5 * 1/6) + (6 * 1/6) = 3.5**  

This means the expected outcome, when rolling the die many times, is 3.5, even though 3.5 is not a possible outcome itself. The expected value gives an average over a large number of trials, reflecting the central tendency of the distribution.'''
```

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt

# Define the range of the discrete uniform distribution
low, high = 1, 6  # For a die roll (1 to 6)

# Generate the discrete uniform distribution
x = np.arange(low, high + 1)  # Possible outcomes (1, 2, 3, 4, 5, 6)
pmf = np.ones_like(x) / len(x)  # Each outcome has equal probability

# Plot the PMF
plt.bar(x, pmf, width=0.5, align='center', color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')

# Show the plot
plt.xticks(x)  # Set x-ticks to match the outcomes
plt.ylim(0, 1)  # Set the y-axis limit to show probabilities from 0 to 1
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

import math

# Function to calculate the PDF of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.
    
    Parameters:
    - x: The outcome of the Bernoulli trial (either 0 or 1).
    - p: The probability of success (the probability of getting a 1).
    
    Returns:
    - Probability for the given outcome (x).
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")
    
    # PDF of a Bernoulli distribution
    if x == 1:
        return p  # Probability of success
    else:
        return 1 - p  # Probability of failure

# Example usage
p = 0.6  # Probability of success (getting 1)
x = 1  # Outcome (1 for success)

# Calculate and print the PDF
pdf_value = bernoulli_pdf(x, p)
print(f"The PDF value for x={x} is: {pdf_value}")
```

```python
#9.  Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    """
    Calculate the Z-scores for a given dataset.
    
    Parameters:
    - data: A list or numpy array of numerical values (dataset).
    
    Returns:
    - z_scores: An array of Z-scores corresponding to the dataset.
    """
    mean = np.mean(data)  # Calculate the mean of the dataset
    std_dev = np.std(data)  # Calculate the standard deviation of the dataset
    
    # Calculate Z-scores
    z_scores = (data - mean) / std_dev
    return z_scores

# Example usage
data = np.random.normal(0, 1, 1000)  # Generate a random dataset from a normal distribution

# Calculate Z-scores for the dataset
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.plot(z_scores, color='purple', alpha=0.7)
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')

# Show the plot
plt.grid(True)
plt.show()
```

```python
#10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the non-normal (uniform) distribution
low, high = 0, 10  # Range for the uniform distribution
size = 1000  # Number of random samples
sample_size = 50  # Size of each sample

# Function to simulate CLT
def central_limit_theorem(low, high, size, sample_size):
    # Generate a non-normal distribution (uniform distribution)
    data = np.random.uniform(low, high, size)
    
    # List to store the sample means
    sample_means = []
    
    # Take repeated samples and calculate the mean of each sample
    for _ in range(size):
        sample = np.random.choice(data, sample_size)
        sample_means.append(np.mean(sample))
    
    return sample_means

# Simulate the Central Limit Theorem
sample_means = central_limit_theorem(low, high, size, sample_size)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')

# Add labels and title
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title(f'Central Limit Theorem: Sample Means from a Uniform Distribution\n(sample_size={sample_size}, size={size})')

# Show the plot
plt.grid(True)
plt.show()
```

```python
#15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution
population_size = 10000  # Total number of data points in the population
sample_size = 50  # Number of elements in each sample
num_samples = 1000  # Number of samples to draw

# Function to simulate CLT with normal distribution
def central_limit_theorem_normal(mean, std_dev, population_size, sample_size, num_samples):
    # Generate a large population from a normal distribution
    population = np.random.normal(mean, std_dev, population_size)
    
    # List to store the sample means
    sample_means = []
    
    # Take multiple samples and calculate their means
    for _ in range(num_samples):
        sample = np.random.choice(population, sample_size)
        sample_means.append(np.mean(sample))
    
    return sample_means

# Simulate the Central Limit Theorem
sample_means = central_limit_theorem_normal(mean, std_dev, population_size, sample_size, num_samples)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title(f'Central Limit Theorem: Sample Means from a Normal Distribution\n(sample_size={sample_size}, num_samples={num_samples})')

# Show the plot
plt.grid(True)
plt.show()
```

```python
#19. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

# Function for Z-test hypothesis testing
def z_test(sample_data, population_mean, population_std, significance_level=0.05):
    """
    Perform a Z-test for hypothesis testing.
    
    Parameters:
    - sample_data: Sample dataset (array-like).
    - population_mean: Population mean under the null hypothesis.
    - population_std: Population standard deviation.
    - significance_level: The significance level (default is 0.05).
    
    Returns:
    - z_statistic: The calculated Z-statistic.
    - p_value: The p-value for the hypothesis test.
    - reject_null: Boolean indicating whether to reject the null hypothesis.
    """
    # Sample statistics
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)
    
    # Z-statistic calculation
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # P-value calculation (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    
    # Compare p-value with significance level to decide whether to reject Hâ
    reject_null = p_value < significance_level
    
    return z_statistic, p_value, reject_null

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=105, scale=10, size=50)  # Sample data (e.g., test scores)

population_mean = 100  # Null hypothesis population mean
population_std = 15  # Known population standard deviation
significance_level = 0.05  # 5% significance level

# Perform the Z-test
z_statistic, p_value, reject_null = z_test(sample_data, population_mean, population_std, significance_level)

# Print results
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis test result
if reject_null:
    print("Reject the null hypothesis (Hâ).")
else:
    print("Fail to reject the null hypothesis (Hâ).")
```

```python
#20. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

# Function to calculate confidence interval
def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given dataset.
    
    Parameters:
    - data: The sample dataset (array-like).
    - confidence_level: The desired confidence level (default is 95%).
    
    Returns:
    - confidence_interval: A tuple containing the lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use sample standard deviation (Bessel's correction)
    sample_size = len(data)
    
    # Z-value for the given confidence level (for large sample sizes, we assume Z-distribution)
    z_value = stats.norm.ppf((1 + confidence_level) / 2)
    
    # Margin of error
    margin_of_error = z_value * (sample_std / np.sqrt(sample_size))
    
    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Sample data (e.g., test scores)

# Calculate the confidence interval for the sample data
confidence_level = 0.95
ci_lower, ci_upper = confidence_interval(sample_data, confidence_level)

# Print the confidence interval
print(f"The {int(confidence_level * 100)}% confidence interval for the sample data is: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for the Poisson distribution
lambda_value = 5  # Mean number of occurrences (lambda)
max_value = 15    # Maximum value to consider for the CDF

# Generate range of x values (possible outcomes)
x_values = np.arange(0, max_value + 1)

# Calculate the CDF for the Poisson distribution
cdf_values = poisson.cdf(x_values, mu=lambda_value)

# Plot the CDF of the Poisson distribution
plt.step(x_values, cdf_values, where='post', label=f'Poisson CDF\n$\lambda$={lambda_value}', color='blue')

# Add titles and labels
plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution')
plt.xlabel('Number of occurrences (k)')
plt.ylabel('Cumulative Probability')
plt.legend()

# Show grid and plot
plt.grid(True)
plt.show()

# Print the CDF values for interpretation
for x, cdf in zip(x_values, cdf_values):
    print(f"P(X <= {x}) = {cdf:.4f}")
```

```python
#27. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
dataset = np.random.normal(loc=50, scale=10, size=1000)

# Calculate the mean and standard deviation of the original dataset
mean = np.mean(dataset)
std_dev = np.std(dataset)

# Z-score normalization
normalized_dataset = (dataset - mean) / std_dev

# Plot original dataset vs. normalized dataset
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot original dataset
axs[0].hist(dataset, bins=30, color='blue', alpha=0.7)
axs[0].set_title('Original Dataset')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')

# Plot normalized dataset
axs[1].hist(normalized_dataset, bins=30, color='orange', alpha=0.7)
axs[1].set_title('Normalized Dataset (Z-score Normalization)')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')

# Show plot
plt.tight_layout()
plt.show()

# Print the mean and standard deviation of the normalized dataset
print(f"Mean of the original dataset: {mean:.2f}")
print(f"Standard deviation of the original dataset: {std_dev:.2f}")
print(f"Mean of the normalized dataset: {np.mean(normalized_dataset):.2f}")
print(f"Standard deviation of the normalized dataset: {np.std(normalized_dataset):.2f}")
```
